```python
import pandas as pd

placement2021_path = "C:\\Users\\Tarun\\Documents\\EDA_Project\\Output\\placement2021.csv"
placement2020_path = "C:\\Users\\Tarun\\Documents\\EDA_Project\\Output\\placement2020.csv"
placement2019_path = "C:\\Users\\Tarun\\Documents\\EDA_Project\\Output\\placement2019.csv"

placement2021_df = pd.read_csv(placement2021_path)
placement2020_df = pd.read_csv(placement2020_path)
placement2019_df = pd.read_csv(placement2019_path)
```

```python
placement2021_df.head
```

```python
institute_summary = placement2021_df.groupby('Institute').agg({
    'No. of students\nplaced': 'sum',
    'No. of students\ngraduating in\nminimum stipulated\ntime': 'sum',
    'No. of students\nselected for Higher\nStudies': 'sum'
}).reset_index()

print(institute_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.barplot(data=institute_summary, x='Institute', y='No. of students\nplaced', color='blue', label='Students Placed')
sns.barplot(data=institute_summary, x='Institute', y='No. of students\ngraduating in\nminimum stipulated\ntime', color='green', label='Students Graduating')
sns.barplot(data=institute_summary, x='Institute', y='No. of students\nselected for Higher\nStudies', color='red', label='Students Higher Studies')

plt.xlabel('Institutes', fontweight='bold')
plt.ylabel('Number of Students')
plt.title('Number of Students Placed, Graduating, and Selected for Higher Studies by Institute 2021')
plt.xticks(rotation=90)

plt.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

placement2021_df['Median salary of\nplaced\ngraduates(Amount in\nRs.)'] = placement2021_df['Median salary of\nplaced\ngraduates(Amount in\nRs.)'].str.replace('[^\d]', '', regex=True).astype(float)

median_salary_by_institute = placement2021_df.groupby('Institute')['Median salary of\nplaced\ngraduates(Amount in\nRs.)'].median()

top_5_institutes = median_salary_by_institute.nlargest(5).index

top_5_df = placement2021_df[placement2021_df['Institute'].isin(top_5_institutes)]

plt.figure(figsize=(12, 6))

sns.barplot(data=top_5_df, x='Institute', y='Median salary of\nplaced\ngraduates(Amount in\nRs.)')

plt.xlabel('Institute', fontweight='bold')
plt.ylabel('Median Salary (Amount in Rs.)')
plt.title('Median Salary of Placed Graduates for Top 5 Institutes (Placement 2021)')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
