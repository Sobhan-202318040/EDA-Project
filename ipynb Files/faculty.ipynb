```python
import pandas as pd

number_of_faculties_df = pd.read_csv(r"C:\Users\Tarun\Documents\EDA_Project\Output\number-of-faculties.csv")

print(number_of_faculties_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.bar(number_of_faculties_df['Institute'], number_of_faculties_df['Number of Faculties'], color='skyblue', width=0.8)

plt.xlabel('Institute', fontsize=12)
plt.ylabel('Number of Faculties', fontsize=12)
plt.title('Number of Faculties in Different Institutes', fontsize=14)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
